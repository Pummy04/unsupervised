```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Import Libraries
from sklearn.cluster import KMeans
from sklearn .datasets import make_blobs
```

```python
x,y=make_blobs(n_samples=300,n_features=2,centers=4,cluster_std=1.5,random_state=42)
```

```python
km=KMeans(n_clusters=4,random_state=42,n_init=10)
```

```python
km.fit(x).score(x)
```

```python
centers=km.cluster_centers_
centers
```

```python
labels=km.labels_
labels
```

```python
plt.scatter(x[:,0],x[:,1],c=labels,s=50,cmap='viridis',marker='o',edgecolor='k')
plt.scatter(centers[:,0],centers[:,1],c='red',marker='*',edgecolor='k',label='centroid')
```
